```python
from scipy import *
import matplotlib.pyplot as plt
import numpy as np
import math
# if using a Jupyter notebook, includes:
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
#%plt.rcParams['figure.fisize']=(6,3)
#%plt.rcParams['figure.dpi']=150

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def func_sin(x,n):
    sin_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i+1)
        denom = math.factorial(2*i+1)
        sin_approx += coef * (num/denom)
    return sin_approx
```

```python
def func_cos(x,n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += coef * (num/denom)
    return cos_approx
```

```python
angles = np.arange(-2*np.pi, 2*np.pi, 0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles, p_cos)

# add lines or between 1 and 6 terms in the Taylor's series
for i in range(1,6):
    t_cos = [func_cos(angle, i) for angle in angles]
    ax.plot(angles, t_cos)
    
ax.set_ylim([-7,4])

# set up legend
legend_lst = ['cos() function']
for i in range(1,6):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=4) 
# loc = location of legend

plt.show()
```

```python
def test(n) :

    angles = np.arange(-4*np.pi, 4*np.pi, 0.1)
    p_cos = np.cos(angles)
    aproxcos = [func_cos(angle,n) for angle in angles]
    p_sin = np.sin(angles)
    aproxsin = [func_sin(angle,n) for angle in angles]
    
    fig, ax = plt.subplots(1,2) # 1 line, 2 columns
    fig.set_size_inches(10,3)
    
    plt.sca(ax[0])
    plt.plot(angles,p_cos)
    plt.plot(angles,aproxcos)
    
    plt.vlines(0,-2.5,2.5)
    plt.hlines(0,min(angles),max(angles))
    plt.legend(['cos() function','Taylor Series - %1.0f term'%n])
    plt.ylim([-2.5,2.5])
    
    
    plt.sca(ax[1])
    plt.plot(angles,p_sin)
    plt.plot(angles,aproxsin)
    
    plt.vlines(0,-2.5,2.5)
    plt.hlines(0,min(angles),max(angles))
    plt.legend(['sin() function','Taylor Series - %1.0f term'%n])
    plt.ylim([-2.5,2.5])
    
    
#interactive
interact(test, n=(1,20))
```
